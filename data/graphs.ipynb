## Initialization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Training Evaluation

### Loss

```python
df = pd.read_csv("train_loss.csv")

fig, ax = plt.subplots(figsize=(9, 6))
window_size = 100

ax.plot(df, label="Actual Loss")
ax.plot(df.rolling(window_size).mean(), label=f"Moving Average Over {window_size} Episodes")

ax.set_xlabel("Episode")
ax.set_ylabel("Loss")
ax.set_title(f"Huber Loss During Training")
ax.legend()

plt.show()
```

### Performance

```python
df = pd.read_csv("train_stat.csv")["score"]

fig, ax = plt.subplots(figsize=(9, 6))
window_size = 10

ax.plot(df, label="Actual Score")
ax.plot(df.rolling(window_size).mean(), label=f"Moving Average Over {window_size} Episodes")

ax.set_xlabel("Episode")
ax.set_ylabel("Score")
ax.set_title(f"Score During Training")
ax.legend()

plt.show()
```

### Random Agent vs RL Agent

```python
df2 = pd.read_csv("random_score.csv")
df3 = pd.read_csv("dqn_score.csv")

fig, ax = plt.subplots(figsize=(9, 6))
window_size = 200

ax.plot(df2.rolling(window_size).mean(), label=f"Random")
ax.plot(df3.rolling(window_size).mean(), label=f"DQN")

ax.set_xlabel("Episode")
ax.set_ylabel("Score")
ax.set_title(f"Score Comparison ({window_size} Episode Moving Avg)")
ax.legend()

plt.show()
```
